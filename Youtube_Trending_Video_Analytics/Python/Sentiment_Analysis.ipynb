## Step 2: Sentiment Classification. üöÄ
Now, let's apply keyword-based sentiment tagging to classify each video as Positive, Neutral, or Negative based on common words in the title.
## Apply Keyword-Based Sentiment Tagging

```python
# Define sentiment-related keywords
positive_words = ["amazing", "best", "great", "success", "love", "top", "winner"]
negative_words = ["fail", "worst", "hate", "problem", "controversy", "disaster"]

# Function to classify sentiment
def classify_sentiment(title):
    title = str(title).lower()
    if any(word in title for word in positive_words):
        return "Positive"
    elif any(word in title for word in negative_words):
        return "Negative"
    else:
        return "Neutral"

# Apply classification to the dataset
df_trending["Sentiment"] = df_trending["title"].apply(classify_sentiment)
```

### NLP would provide a much more accurate sentiment analysis than our keyword-based approach. üöÄ
The current method relies on predefined words, which can be limiting. TextBlob or NLTK can analyze context, tone, and phrasing, leading to better sentiment classification.

```python
from textblob import TextBlob
```

### Apply Sentiment Analysis
TextBlob assigns polarity scores (-1 to +1) to text:
- Positive (>0) ‚Üí Optimistic, happy, or engaging content.
- Neutral (=0) ‚Üí Objective or factual titles.
- Negative (<0) ‚Üí Critical or controversial topics.

```python
df_trending["Sentiment_Score"] = df_trending["title"].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

‚úî This assigns a score based on the tone of each title.

```python
## Categorize Sentiment
We now classify sentiment based on score:
```

```python
def categorize_sentiment(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df_trending["Sentiment"] = df_trending["Sentiment_Score"].apply(categorize_sentiment)
```

```python
‚úî This ensures videos are categorized accurately based on NLP analysis.
```

### Analyze Sentiment Trends by Category

```python
import pandas as pd

# Group by category and calculate average sentiment score
category_sentiment = df_trending.groupby("category_name")["Sentiment_Score"].mean().reset_index()

# Sort results for better comparison
category_sentiment = category_sentiment.sort_values(by="Sentiment_Score", ascending=False)

# Display top categories by sentiment
print(category_sentiment.head(10))
```

```python
## üîç Key Observations
‚úî Nonprofits & Activism has the highest average sentiment score (0.1767) ‚Üí Titles in this category are generally uplifting or positive.
‚úî Travel & Events ranks second (0.1486) ‚Üí Suggests travel-related content often has exciting or aspirational themes.
‚úî Cars & Vehicles / Autos & Vehicles (0.1042) ‚Üí Indicates that automotive content typically leans positive.
‚úî Sports (0.0938) ‚Üí Shows that sports-related trending videos evoke slightly positive sentiment, possibly due to enthusiasm and game highlights.
‚úî Music (0.0837) ‚Üí Generally positive, but not overwhelmingly so‚Äîmusic titles may be neutral or descriptive instead of emotional.
‚úî People & Blogs (0.0698) ‚Üí Falls lower on the list, meaning personal storytelling and vlogs might have a mix of neutral or slightly positive sentiment.
```

## Merge Sentiment Scores with youtube_trending_with_categories.csv
Now, let‚Äôs combine the sentiment results into the dataset:

```python
import pandas as pd

# Load main dataset
df = pd.read_csv(r"C:\Users\91961\Downloads\youtube_trending_with_categories.csv")

# Load sentiment results
sentiment_df = pd.read_csv(r"C:\Users\91961\Downloads\sentiment_results.csv")

# Merge sentiment scores using video_id as the key
df = df.merge(sentiment_df[['video_id', 'sentiment_score']], on="video_id", how="left")

# Remove any duplicate entries
df = df.drop_duplicates(subset="video_id")

# Save the updated dataset
df.to_csv("youtube_trending_with_sentiment.csv", index=False)
print("‚úÖ Sentiment scores successfully integrated into the dataset!")
```
